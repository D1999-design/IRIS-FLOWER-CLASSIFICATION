```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stop_words=stopwords.words('english')
print(stop_words)
```

```python
#Initializing Dataset
```

```python
df = pd.read_csv("/content/spam.csv",encoding='latin-1')


# Displaying the first few rows of the dataset
df.head()
```

```python
df
```

```python
#Data Cleaning
```

```python
data=df.where((pd.notnull(df)),'')
```

```python
data
```

```python
data.head()
```

```python
data.describe()
```

```python
#INITIALIZING HAM AS 1 AND SPAM AS 0
```

```python
data.loc[data['v1']=='spam','v1',]=0
data.loc[data['v1']=='ham','v1',]=1
```

```python
x=data['v2']
y=data['v1']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#Extracting Features
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words="english", lowercase=True)
```

```python
x_train_feature = feature_extraction.fit_transform(x_train)
```

```python
x_test_feature = feature_extraction.transform(x_test)
```

```python
y_train_feature = y_train.astype('int')
```

```python
y_test_feature = y_test.astype('int')
```

```python
#Training the model using Logistic Regression
```

```python
model=LogisticRegression()
model.fit(x_train_feature,y_train_feature)
```

```python
pred=model.predict(x_train_feature)
print(accuracy_score(y_train_feature,pred)*100)
```

```python
pred2=model.predict(x_test_feature)
print(accuracy_score(y_test_feature,pred2)*100)
```

```python
inp=["Hey!! Mr Rithvik you have been elected as our next ceo"]
```

```python
input_features=feature_extraction.transform(inp)
prediction=model.predict(input_features)
print(prediction)
```

```python
if(prediction==1):
    print('HAM MAIL')
else:
    print("SPAM MAIL")
```
